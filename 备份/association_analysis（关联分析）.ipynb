```python
#获取数据源
from dota2_api import get_api_json
import pickle

with open('hero_name.pickle', 'rb') as handle:  
    hero_name = pickle.load(handle)

sql = """SELECT
matches.match_id,
matches.start_time,
matches.radiant_win,
leagues.name leaguename,
matches.picks_bans
FROM matches
JOIN leagues using(leagueid)
where leagueid = 9943 and to_timestamp(start_time) >= now() - interval '1 month'
ORDER BY start_time desc"""

rows = get_api_json('https://api.opendota.com/api/explorer?sql={}'.format(sql))['rows']
all_picks = []
for row in rows:
    radiant_picks = [ bp['hero_id'] for bp in row['picks_bans'] if bp['team'] == 0 and bp['is_pick']]
    dire_picks = [ bp['hero_id'] for bp in row['picks_bans'] if bp['team'] == 1 and bp['is_pick']]
    all_picks.append(radiant_picks)
    all_picks.append(dire_picks)
all_picks
```

```python
len(all_picks)
```

```python
all_picks_hero = [ [ hero_name[hi][1] for hi in pick] for pick in all_picks ]
all_picks_hero
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori


te = TransactionEncoder()
te_ary = te.fit(all_picks_hero).transform(all_picks_hero)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.015, use_colnames=True)

frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0)
rules["antecedant_len"] = rules["antecedants"].apply(lambda x: len(x))
# rules

select_rules =  rules[ (rules['antecedant_len'] >= 1) &
       (rules['confidence'] > 0.4) &
       (rules['lift'] > 1.2) ]

select_rules
```

```python
rows = get_api_json('https://api.opendota.com/api/explorer?sql={}'.format(sql))['rows']
all_picks_wl = []
for row in rows:
    radiant_picks = [ hero_name[bp['hero_id']][1] for bp in row['picks_bans'] if bp['team'] == 0 and bp['is_pick']]
    dire_picks = [ hero_name[bp['hero_id']][1] for bp in row['picks_bans'] if bp['team'] == 1 and bp['is_pick']]
    all_picks_wl.append([radiant_picks, row['radiant_win']])
    all_picks_wl.append([dire_picks, not row['radiant_win']])
all_picks_wl
```

```python
def get_hc_win_rate(all_picks_wl, hc):
    count = 0
    win_count = 0 
    for picks_wl in all_picks_wl:        
        if all(h in picks_wl[0] for h in hc):
            count += 1
            if picks_wl[1] is True:
                win_count += 1
    return win_count/count

# get_hc_win_rate(all_picks_wl, ['艾欧', '灰烬之灵'])

for i, row in select_rules.iterrows():
    hero_combination = []
    hero_combination += list(row['antecedants'])
    hero_combination += list(row['consequents'])
    select_rules.loc[i,'胜率'] = '{:.1%}'.format(get_hc_win_rate(all_picks_wl, hero_combination))
    select_rules.loc[i,'出场次数'] = 206 * row['support']
    
    
pd.options.display.float_format = '{:.3f}'.format    
select_rules['出场次数'] = select_rules['出场次数'].map('{:.0f}'.format)
select_rules[['antecedants', 'consequents', 'support', 'confidence', 'lift', '胜率', '出场次数']]
```
